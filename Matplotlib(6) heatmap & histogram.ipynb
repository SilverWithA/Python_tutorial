```python
import matplotlib.pyplot as plt
import pandas as pd
```

## 히트맵

색을 이용해 값의 크기를 표현하여 비교하는 그래프

``plt.pcolor``(data)

여기서 데이터는 2차원 데이터여야 한다.

```python
import numpy as np
arr = np.random.standard_normal((4,4)) 
df = pd.DataFrame(arr)
df
```

* np.random.standard_normal
참고: https://codetorial.net/numpy/functions/numpy_random_standard_normal.html


표준정규분포(standard normal distribution)에서 샘플링한 난수를 반환

```python
# 히트맵 그리기

plt.pcolor(df,cmap = 'cool') #컬러맵 cool 사용
plt.colorbar()
```

### seaborn으로 히트맵 그리기

* heatmap(data)<br>

``annot``=True : 수치 표시<br>
``fmt``='d' : 정수로 표시/'f'=실수로 표시<br>

```python
import seaborn as sns
```

```python
sns.heatmap(df,cmap='Blues',annot = True, fmt ='f')
```

## 히스토그램

도수분포를 나타내는 그래프
x축으로 데이터 값의 구간, y축은 구간별 값의 개수(count)

* plt.``hist``(data, bins = 구간의 개수, cumulative = 누적히스토그램 여부, range(min,max))

``cumulative`` = 누적 히스토그램

``range`` = 구간의 범위(x축)

``density`` = 밀도. 도수의 총 개수를 나눈 수치를 y축에 표시

``orientation``= horizontal = 가로 히스토그램

* 선만 표시 <b>`histtype='step'`</b>

`rwidth=막대폭(0~1)` <br>
`color=막대색` <br>
`alpha=투명도`<br>
`edgecolor(ec)=선색`<br>
`linewidth(lw)=선두께`<br> `linestyle(ls)=선스타일`<br>
`hatch=패턴`<br>


```python
arr1 = np.random.standard_normal(100)
df = pd.DataFrame(arr1)

plt.hist(df,bins=10,color='red',alpha = 0.5)
```

```python
plt.hist(df,bins=10,cumulative= True)
```
