## [1] Simple line plot & basic skills

```python
# 그래프로 확인
x = ['봄','여름','가을','겨울']
y = [19,30,15,-10]

# 그래프 그리기
plt.plot(x,y)
```

### 1. 색상 지정 및 옵션 바꾸기

plt.plot(x,y, color = 색상, marker = 마커모양, ls = 선스타일, lw= 선굵기)

* 색상이름 약자(blue:b, green:g, red:r, cyan:c, magenta:m, yellow:y, black:k, white:w)
* 마커종류 : * . , o v ^ < > 1 2 3 4 s p * h H + x D d
* 선 스타일


``'-'``          solid line style    
``'--'``         dashed line style    
``'-.'``         dash-dot line style        
``':'``          dotted line style    

```python
plt.plot(x,y,color='#A566FF', marker='D', ms='10', mec='b', mew='3', mfc='y'
        ,ls=':', lw=3)
```

## [2] bar chart

수량시각화에 사용

* plt.bar()

```python
blood_type=['A','B','O','AB']
count=[45,35,15,5]
plt.bar(blood_type,count)
```

```python
# 가로 그래프
plt.barh(blood_type,count)
```

## [3] pie chart

비율 시각화에 유용

* plt.pie(data, labels,autopct, shadow, explode,counterclock,startangle)
* 옵션

`` labels ``  : 각 부채꼴에 표시될 항목의 이름

`` autopct ``: 각 부채꼴에 표시될 숫자의 형식. '%0.1f'가 들어가면 소수점 첫째 자리까지 표시

```python
plt.figure(figsize=(3,3))
plt.pie(count, labels=blood_type, autopct = '%0.0f%%') # 소수점 0의 자리까지-> 정수만 출력
plt.show()
```

## [4] histogram

분포의 시각화

* plt.hist(data, bins=나눌 계급의 개수)

```python
import numpy as np
data = np.random.randint(1,101,1000)

plt.hist(data, bins=50)
plt.show()
```

## [5] box plot

이상치 제거에 활용

```python
plt.boxplot(data)
plt.show()
```

## [6] sacatter

산점도. 데이터이 관계성을 시각화할 때 사용

* plt.sacatter(x,y,s = 마커크기, r= 마커색상)

```python
# 총금액과 팁과의 관계 살펴보기

import seaborn as sns
tips = sns.load_dataset('tips')[['total_bill','tip']]
tips
```

```python
plt.scatter(tips['total_bill'], tips['tip'])
plt.show()
```
