```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False
```

# 카운트플롯
* 데이터의 갯수를 카운트하여 시각화 
* `sns.countplot(data=데이터프레임, x=컬럼)`<br>


```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# 연령별 승선인원 카운트하여 시각화

plt.figure(figsize=(15,4))
sns.countplot(titanic, x='age')
plt.xticks(rotation = 90)
plt.show()
```

```python
# 1. 남자/여자/어린이 승선인원 시각화
titanic['who'].unique()
```

```python
sns.countplot(titanic, x= 'who')
```

```python
# 2. 남자/여자/어린이 별 생존여부 

sns.countplot(titanic, x= 'who', hue= 'alive')
```

# 히스토그램
* `sns.displot(data=데이터프레임, x=컬럼)`

```python
iris = sns.load_dataset('iris')
iris.head(1)
```

```python
sns.displot(iris, x='petal_length',hue='species', bins=20,
           rug =True, # 러그플롯추가
           kde= True) # 선추가
```

```python
sns.displot(data=iris, x='petal_length', 
            rug=True, hue='species', kind='kde')
```

# 상자수염그래프, 바이올린플롯, 스트립플롯, 스웜플롯
* sns.boxplot(data=데이터프레임)
* sns.violinplot(data=데이터프레임)
* sns.stripplot(data=데이터프레임)
* sns.swarmplot(data=데이터프레임)

```python
plt.figure(figsize=(12,8))
plt.subplot(221)

sns.boxplot(iris)

plt.subplot(222)
sns.violinplot(iris)

plt.subplot(223)
sns.stripplot(iris)

plt.subplot(224)
sns.swarmplot(iris)

plt.show()

```
