## 산점도

* ``plt.scatter``(x,y,s=점의 크기, c= 점의 색상)
* ``cmap = 컬러맵`` 사용 가능

plt.colorbar(): 컬러바 표시

plt.colormaps() # 컬러맵 종류 보기

```python
# 컬러맵
x = [1,5,6,9,10]
y = [1,5,3,9,7]

plt.scatter(x,y,c=y, cmap='viridis')
plt.colorbar()
```

```python
plt.colormaps()[:10] # 컬러맵 종류 보기
```

### 산점도 실습

```python
import seaborn as sns
tips = sns.load_dataset('tips')
tips.head()
```

```python
# 성별에 따른 color 컬럼 추가


def set_color(x):
    if x=='Male':
        return 'blue'
    elif x=='Female':
        return 'red'
    
tips['color'] = tips['sex'].apply(set_color)   
```

```python
plt.scatter(tips['total_bill'], tips['tip'], 
            s=tips['size']*50, alpha=0.5, c=tips['color'])
```
