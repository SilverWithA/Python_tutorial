### 1. 여러 그래프 그리기

#### (1) 하나의 그래프 창에 여러 그래프 그리기

* ``plt.figure(n)`` : n번째 새로운 그래프 창을 생성함
* ``plt.clf()`` : 해당 그래프창에 그려진 모듬 그래프를 지움

```python
x = ['봄','여름','가을','겨울']
y = [19,30,15,-10]

plt.figure(figsize =(3,3))
plt.plot(x,y)

plt.figure() # 새로운 그래프 창 생성

plt.figure(figsize =(3,3))
plt.plot(x,y,"y") 
plt.show()
```

#### (2) 하나의 그래프 창에 하이 그래프 영역 나눠 그리기


* ``plt.figure``(figsize = (x사이즈, y사이즈),facecolor= 색상) - 전체 그래프의 크기


fig = plt.figure() 를 통해 fig 객체 지정

fig.suptitle("전체 그래프의 제목")

fig.tight_layout() 그래프의 너비와 간격을 최적화

* ``plt.subplot``(m,n,p)

m x n 행렬로 이뤄진 하위 그래프 중 p번째 위치에 그래프가 그려지도록 지정함.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,10,0.1)
y1 = 0.3*(x-5)**2 +1
y2 = 1.5*x+3
y3 = np.sin(x)**2
y4 = 10 * np.exp(-x)+1

fig = plt.figure(figsize=(7,7),facecolor = "ivory")
plt.subplot(2,2,1) # p = 1
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)

fig.suptitle("< Total Graph >")
fig.tight_layout()
```

### 2.라벨, 제목, 격자, 범례, 문자열 표시

#### (1) 축 라벨과 그래프 제목

x축, y축에 이름 달기

* plt.xlabel("x축이름")
* plt.ylabel("y축이름")


* loc : 위치('left','center','right' / 'bottom', 'center','top')
* labelpad : 레이블과 그래프와의 간격
* fontsize : 폰트사이즈
* fontfamily : 폰트



그래프 제목

* plt.title("그래프의 이름")

```python
x = np.arange(-4,5,0.5)
y = 2*x**3

plt.figure(figsize = (5,5))
plt.plot(x,y)
plt.xlabel("x축",loc ='left')
plt.ylabel("y축",loc = 'top')
plt.title("기본 그래프")
plt.grid(False) # axis =x 로 지정해줄시 x축으로만 그리드 생성 가능/ color 지정가능
plt.show()
```

#### (2) 범례지정

* ``plt.legend``[범례이름 리스트],loc=범례 박스 위치 )

```python
x = np.arange(0,5,1)
y1= x+1
y2 =x+2
y3 =x+3
y4 =x+4


plt.figure(figsize = (4,4))
plt.plot(x,y1,'r',
        x,y2,'y',
        x,y3,'b',
        x,y4,'w')

plt.legend(['d1','d2','d3','d4'],loc='upper right' )
plt.show()
```

#### (3) 문자열 추가
* ``plt.text``(x좌표위치, y좌표위치, "문자열")

```python
plt.figure(figsize = (4,4))
plt.plot(x,y1,'r',
        x,y2,'y',
        x,y3,'b',
        x,y4,'w')


plt.legend(['d1','d2','d3','d4'],loc='upper right' )
plt.text(2,2,"ㅇㅇ")
plt.show()
```
